```python
# Found out mse, rmse, mpe. 
# Plotted the scaled predicted volume and true volume of apple for comparison.
# Plotted frequence and absolute error
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../Data/scaled_predicted_volume.csv')
```

```python
mse = ((df['absolute_error']) ** 2).mean()
rmse =  mse ** 0.5
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (MSE):", rmse)
```

```python
mean_percentage_error = (df['percentage_error']).mean()
print("Mean Percentage Error (MPE):", mean_percentage_error)
```

```python
plt.figure(figsize=(20, 7.5))
plt.plot(df.index, df['true_volume'], color='green', label='True Volume')
plt.plot(df.index, df['scaled_predicted_volume'], color='blue', label='Predicted Volume')
plt.xlabel('Apple label')
plt.ylabel('Volume')
plt.title('Plot of Volume vs Apple Label')
plt.xlim(0, 200) 
plt.ylim(0, 400)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.hist(df['absolute_error'], bins=50, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Absolute Error')
plt.ylabel('Frequency')
plt.title('Frequency Plot of Absolute Error')
plt.grid(True)
plt.show()
```
